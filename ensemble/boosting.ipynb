```python
import pandas as pd
import numpy as np
df=pd.read_csv("C:\\Users\\dobil\\OneDrive\\Documents\\Tekworks\\Social_Network_Ads.csv")
df.head()
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,[2,3]].values #age,estimated salary
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
s=StandardScaler()
x_train_scaled=s.fit_transform(x_train)
x_test_scaled=s.transform(x_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(max_depth=1)
classifier1=AdaBoostClassifier(
    estimator=model,n_estimators=50,learning_rate=1.0,random_state=42
)
# classifier2=GradientBoostingClassifier(
#     n_estimators=100,learning_rate=1.0,max_depth=3,random_state=42
# )
classifier1.fit(x_train_scaled,y_train)
```

```python
classifier2=GradientBoostingClassifier(
    n_estimators=100,learning_rate=0.1,max_depth=3,random_state=42
)
classifier2.fit(x_train_scaled,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred1=classifier1.predict(x_test_scaled)
y_pred2=classifier2.predict(x_test_scaled)
print("Accuracy Score1;",accuracy_score(y_test,y_pred1))
print("Accuracy Score2:",accuracy_score(y_test,y_pred2))
```
