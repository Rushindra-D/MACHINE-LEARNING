```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(1,21).reshape(-1,1)
y=np.array([
    10,12,15,10,20,25,30,28,35,40,
    45,42,50,55,60,58,65,70,72,75
])
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2
,random_state=42)
dt_reg=DecisionTreeRegressor(random_state=42)
dt_reg.fit(x_train,y_train)
y_pred=dt_reg.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg=RandomForestRegressor(n_estimators=100,random_state=42)
rf_reg.fit(x_train,y_train)
y_rf_pred=rf_reg.predict(x_test)
mse_rf=mean_squared_error(y_test,y_rf_pred)
r2_rf=r2_score(y_test,y_rf_pred)
print(f"Random Forest - Mean Squared Error: {mse_rf}")
print(f"Random Forest - R^2 Score: {r2_rf}")
```

```python
plt.scatter(x,y,color='red')
x_grid=np.linspace(1,20,100).reshape(-1,1)
y_grid_pred=dt_reg.predict(x_grid)
plt.plot(x_grid,y_grid_pred,color='blue',label='Decision Tree Prediction')
y_rf_grid_pred=rf_reg.predict(x_grid)
plt.plot(x_grid,y_rf_grid_pred,color='green',label='Random Forest Prediction')
plt.xlabel('Units Consumed')
plt.ylabel('Electricity Bill Amount')
plt.title('Electricity Bill Prediction using Decision Tree and Random Forest Regression')
plt.legend()
plt.show()
```
