 # Decision tree regression
 

```python
import numpy as np
import matplotlib.pyplot as plt
X=np.array([50,120,180,250,320,380]).reshape(-1,1)
y=np.array([500,1000,1500,1800,2400,3000])

```

```python
from sklearn.tree import DecisionTreeRegressor
dt_reg=DecisionTreeRegressor(max_depth=3, random_state=42)
dt_reg.fit(X,y)
```

```python
X_grid=np.linspace(0,400,200).reshape(-1,1)
y_pred=dt_reg.predict(X_grid)
plt.scatter(X,y,color='red')
plt.plot(X_grid,y_pred,color='blue')
plt.xlabel('Units Consumed')
plt.ylabel('Electricity Bill')
plt.title('Decision Tree Regression - Electricity Bill Prediction')
plt.show()

```

```python
u=float(input("Enter the number of units consumed: "))
bill=dt_reg.predict(np.array([[u]]))
print(f"The predicted electricity bill for {u} units is: {bill[0]}")
```
