# Polynomial Regression

```python
import pandas as pd
import numpy as np
df=pd.read_csv("C:\\Users\\dobil\\OneDrive\\Desktop\\MACHINE-LEARNING\\Position_Salaries.csv")
```

```python
x=df.iloc[:,1:-1].values #independent variables
y=df.iloc[:,-1].values #dependent variable
df.head()
```

# Training Model 

```python

from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x, y)
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(x)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
 
print("✅ Models Trained Successfully!")
```

# Visualization

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color='black')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
#polynomial 
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg_2.predict(X_poly),color='black')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

# Prediction

```python
Linear_prediction=lin_reg.predict([[6.5]])
print("Linear Prediction for 6.5:", Linear_prediction)
Polynomial_prediction=lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
print("Polynomial Prediction for 6.5:", Polynomial_prediction)
```

# Regression metrics

```python
from sklearn.metrics import mean_squared_error, r2_score,root_mean_squared_error
y_pred=lin_reg.predict(x)
mse=mean_squared_error(y, y_pred)
r2=r2_score(y, y_pred)
rsme=root_mean_squared_error(y, y_pred)
print(f"Linear Regression MSE: {mse}, R2: {r2}, RSME: {rsme}")
y_pred_poly=lin_reg_2.predict(X_poly)
mse_poly=mean_squared_error(y, y_pred_poly)
r2_poly=r2_score(y, y_pred_poly)
rsme_poly=root_mean_squared_error(y, y_pred_poly)
print(f"Polynomial Regression MSE: {mse_poly}, R2: {r2_poly}, RSME: {rsme_poly}")
```
